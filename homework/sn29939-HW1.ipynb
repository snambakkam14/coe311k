```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def pi_approx(samples):
    counter = 0
    for x in range(0,samples):
        a = random.uniform(-1,1)
        b = random.uniform(-1,1)
        if (a**2 + b**2 < 1):
            counter = counter + 1
    num = counter/samples
    approx = num * 4
    return approx

 
```

```python
error = []

for i in range(10,10000,20):
    mc_sim = pi_approx(i)
    err = np.abs(np.pi - mc_sim)
    error.append(err)
    print(f"Samples={i}, Error={err:.15f}")
    
```

```python
samp = np.arange(10,10000,20)

plt.figure()
plt.plot(samp,error)

```
