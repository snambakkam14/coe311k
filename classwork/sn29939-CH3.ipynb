```python
import random
import math
import numpy as np
import matplotlib.pyplot as plt

def pi_approx(samples):
    counter = 0
    for x in range(0,samples):
        a = random.uniform(-1,1)
        b = random.uniform(-1,1)
        if (a**2 + b**2 < 1):
            counter = counter + 1
    num = counter/samples
    approx = num * 4
    return approx

x=0
true = []
relative = []
axis = np.linspace(0,100,100)

while (x<100):
    b = pi_approx(1000)
    true_error = np.abs(b - np.pi)
    relative_error = np.abs(b-np.pi)/b
    true.append(true_error)
    relative.append(relative_error)
    x = x + 1
    
    
plt.figure()
plt.plot(axis,true,'r')
plt.show()
plt.figure()
plt.plot(axis,relative,'b')
plt.show()

    
    
    
```
