```python
import random
import math

def pi_approx(samples):
    counter = 0
    for x in range(0,samples):
        a = random.uniform(-1,1)
        b = random.uniform(-1,1)
        if (a**2 + b**2 < 1):
            counter = counter + 1
    print(counter)
    num = counter/samples
    approx = num * 4
    return approx

print(pi_approx(300000))
```
