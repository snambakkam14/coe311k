```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x))
plt.show()
```

```python
x = np.linspace(0,2*np.pi,100)
plt.plot(x,np.sin(x),x,np.sin(2*x))
plt.show()
```

```python

y = np.sin(x)
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.scatter(x,y)
plt.subplot(2,2,3)
plt.hist(y,50)
plt.subplot(2,2,4)
plt.plot(x,y**2,'y')
plt.show()
```
